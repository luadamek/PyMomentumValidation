```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib as mpl
import ipyplot

def get_fig_ax(image):
    dpi = mpl.rcParams['figure.dpi']
    img = mpimg.imread(image)
    height, width, depth = img.shape
    figsize = width / float(dpi), height / float(dpi)

    fig = plt.figure(frameon=False, figsize = figsize)

    ax = plt.Axes(fig, [0., 0., 1., 1.])
    ax.set_axis_off()
    fig.add_axes(ax)
    #fig.set_size_inches(8,6)
    
    ax.imshow(img, aspect='auto')
    
    return fig,ax

fig, ax = get_fig_ax('/project/def-psavard/ladamek/pymomentumvaldiation/SolutionHistograms_Dec3_20_0000_100_0000_MS/Canvas_RemainingBias_Local_MS.png')
plt.show()
fig2, ax2 = get_fig_ax('/project/def-psavard/ladamek/pymomentumvaldiation/SolutionHistograms_Dec3_12_0000_100_0000_MS/Canvas_RemainingBias_Local_MS.png')
plt.show()
```

```python
w=10
h=10
fig=plt.figure(figsize=(8, 8))
name_base = '/project/def-psavard/ladamek/pymomentumvaldiation/SolutionHistograms_Dec3_{}_{}_{loc}/Canvas_RemainingBias_Local_{loc}.png'
columns = ["12_0000", "16_0000", "20_0000"]
rows = ["-1_0000", "100_0000"]
from PIL import Image
for loc in ["ID", "MS"]:
    for r in rows:
        images = []
        labels = []
        for c in columns:
            images.append(Image.open(name_base.format(c, r, loc=loc)))
            labels.append("pt_thresh = {} GeV, r = {} GeV, {}".format(r.replace("_0000", ""), c.replace("_0000", ""), loc))

        ipyplot.plot_images(images, labels, img_width = 300)
```
